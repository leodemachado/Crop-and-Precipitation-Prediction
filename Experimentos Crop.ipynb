### Loading City data

```python
def to_float(num):
    if num == '-':
        new_num = np.nan
        return new_num
    else:
        new_num = float(num)
        return new_num
```

```python
cidade = cidade.applymap(to_float)
cidade.dropna()
```

```python
anomalias = ['el', 'NTA', 'TNA', 'TSA']
clima = ['insolacao', 'pressao', 'temp_condensada', 'humidade', 'precipitação']
```

```python
def is_in(feat=None, tolkens=None):
    check=None
    for i in tolkens:
        if i in feat:
            return True
    return False
```

```python
cidade_nome = 'taua_total'
file = "Experimentos_crop/" + cidade_nome +".csv"
target = "crop_" + cidade_nome
```

```python
cidade = pd.read_csv(file)
#del fortaleza['Unnamed: 0']
cidade.head()
```

### Slicing the data set to remove specific features

```python
cidade.columns
```

```python
df_reduced = cidade[[f for f in cidade.columns if not is_in(f, anomalias)]]
df_reduced.columns
```
